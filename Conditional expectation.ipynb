## 1. Probability space $(\Omega, \mathfrak{F}, P)$ where $\mathfrak{F} = \mathfrak{P}(\Omega)$
[cite_start]Since $\mathfrak{F}$ is the power set of a 4-element set, it contains $2^4 = 16$ events:
$$\mathfrak{F} = \{\emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1, \omega_2\}, \{\omega_1, \omega_3\}, \{\omega_1, \omega_4\}, \{\omega_2, \omega_3\}, \{\omega_2, \omega_4\}, \{\omega_3, \omega_4\}, \{\omega_1, \omega_2, \omega_3\}, \{\omega_1, \omega_2, \omega_4\}, \{\omega_1, \omega_3, \omega_4\}, \{\omega_2, \omega_3, \omega_4\}, \Omega\}$$

## 2. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ partitions the sample space into two sets: $\{Y=0\} = \{\omega_1, \omega_2\}$ and $\{Y=1\} = \{\omega_3, \omega_4\}$. Thus:
$$\sigma(Y) = \{\emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega\}$$

## 3. $\sigma(Y)$-measurability of $\mathbb{E}[X|Y]$
By definition, the conditional expectation $\mathbb{E}[X|Y]$ is a random variable that is measurable with respect to the conditioning information $\sigma(Y)$.  The smallest sub-blocks in $\sigma(Y)$ are $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$.For a function to be measurable relative to this topology, it cannot distinguish between individual points inside these partition setsThus, it must take uniform constant values across them

## 4. Explicit Calculation of $\mathbb{E}[X|Y]$
* For $Y=0$ (set $A_0 = \{\omega_1, \omega_2\}$):
  $$\mathbb{E}[X|Y=0] = \frac{\mathbb{E}[X \cdot \mathbb{1}_{\{\omega_1, \omega_2\}}]}{P(\{\omega_1, \omega_2\})} = \frac{0 \cdot 0.50 + 1 \cdot 0.10}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$
* For $Y=1$ (set $A_1 = \{\omega_3, \omega_4\}$):
  $$\mathbb{E}[X|Y=1] = \frac{\mathbb{E}[X \cdot \mathbb{1}_{\{\omega_3, \omega_4\}}]}{P(\{\omega_3, \omega_4\})} = \frac{0 \cdot 0.20 + 1 \cdot 0.20}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$

Thus, $\mathbb{E}[X|Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$.

## 5. Verifying the Defining Integral Property
We check $\int_A \mathbb{E}[X|Y] dP = \int_A X dP$ for all elements in $\sigma(Y)$:
* For $A = \emptyset$: Both sides equal $0$.
* For $A = \{\omega_1, \omega_2\}$:
  $$\text{LHS} = \frac{1}{6} \cdot P(\{\omega_1, \omega_2\}) = \frac{1}{6}(0.6) = 0.1, \quad \text{RHS} = X(\omega_1)P(\omega_1) + X(\omega_2)P(\omega_2) = 0 + 1(0.1) = 0.1$$
* For $A = \{\omega_3, \omega_4\}$:
  $$\text{LHS} = \frac{1}{2} \cdot P(\{\omega_3, \omega_4\}) = \frac{1}{2}(0.4) = 0.2, \quad \text{RHS} = X(\omega_3)P(\omega_3) + X(\omega_4)P(\omega_4) = 0 + 1(0.2) = 0.2$$
* For $A = \Omega$: $\text{LHS} = 0.1 + 0.2 = 0.3$, $\text{RHS} = \mathbb{E}[X] = 0.1 + 0.2 = 0.3$.

## 6 & 7. Conditional PMF identity
$$p_{X|Y}(x|0) = \begin{cases} 5/6 & x=0 \\ 1/6 & x=1 \end{cases}, \quad p_{X|Y}(x|1) = \begin{cases} 1/2 & x=0 \\ 1/2 & x=1 \end{cases}$$
[cite_start]Summing $\sum x \cdot p_{X|Y}(x|y)$ yields $1\cdot(1/6) = 1/6$ for $y=0$ and $1\cdot(1/2) = 1/2$ for $y=1$ , matching $\mathbb{E}[X|Y]$.

## 8 & 9. MSE and Law of Total Variance
* **With information $Y$:** $\text{MSE} = \mathbb{E}[(X - \mathbb{E}[X|Y])^2] = 0.6 \cdot (\frac{5}{6}(0-\frac{1}{6})^2 + \frac{1}{6}(1-\frac{1}{6})^2) + 0.4 \cdot (\frac{1}{2}(0-\frac{1}{2})^2 + \frac{1}{2}(1-\frac{1}{2})^2) = 0.6(\frac{5}{36}) + 0.4(\frac{1}{4}) = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.1833$.
* **Without information $Y$ (Using $\mathbb{E}[X] = 0.3$):** $\text{MSE}_{\text{global}} = \text{Var}(X) = 0.3 \times 0.7 = 0.21$.
* **Relationship:** According to the Law of Total Variance: $\text{Var}(X) = \mathbb{E}[\text{Var}(X|Y)] + \text{Var}(\mathbb{E}[X|Y])$Ignoring $Y$ discards the explained variance $\text{Var}(\mathbb{E}[X|Y])$, increasing the error exactly by that amount.

## 10. Information Theory Analysis
* $H(X) = -0.3\log_2(0.3) - 0.7\log_2(0.7) \approx 0.8813$ bits.
*$H(X|Y) = 0.6 \cdot H(X|Y=0) + 0.4 \cdot H(X|Y=1) = 0.6(- \frac{1}{6}\log_2\frac{1}{6} - \frac{5}{6}\log_2\frac{5}{6}) + 0.4(1) \approx 0.6(0.6500) + 0.4 = 0.7900$ bits.
$I(X;Y) = H(X) - H(X|Y) = 0.8813 - 0.7900 = 0.0913$ bits[cite: 48].
* **Uncertainty Reduction:** $\frac{I(X;Y)}{H(X)} \times 100\% = \frac{0.0913}{0.8813} \times 100\% \approx 10.36\%$

```python
import numpy as np

# Let's set up the programmatic verification of the discrete outcomes
P = np.array([0.50, 0.10, 0.20, 0.20]) # omega_1, omega_2, omega_3, omega_4
X = np.array([0, 1, 0, 1])
Y = np.array([0, 0, 1, 1])

# Computing conditional expectation values
y_u = np.unique(Y)
E_X_given_Y = np.zeros_like(X, dtype=float)

for y in y_u:
    mask = (Y == y)
    prob_cond = P[mask] / np.sum(P[mask])
    val = np.sum(X[mask] * prob_cond)
    E_X_given_Y[mask] = val

print(f"Calculated E[X|Y] array mapping: {E_X_given_Y}")

# Verify Integral Property
for y in y_u:
    mask = (Y == y)
    lhs = np.sum(E_X_given_Y[mask] * P[mask])
    rhs = np.sum(X[mask] * P[mask])
    print(f"Set Y={y} verification -> LHS (Expectation Integral): {lhs:.4f} | RHS (Target Integral): {rhs:.4f}")

# MSE Analysis
mse_with_y = np.sum(((X - E_X_given_Y) ** 2) * P)
global_mean = np.sum(X * P)
mse_without_y = np.sum(((X - global_mean) ** 2) * P)

print(f"\nMSE with Risk Class info: {mse_with_y:.4f}")
print(f"MSE using Global Average: {mse_without_y:.4f}")

# Entropy Calculation
def entropy(probs):
    probs = probs[probs > 0]
    return -np.sum(probs * np.log2(probs))

H_X = entropy(np.array([np.sum(P[X==0]), np.sum(P[X==1])]))
H_X_Y0 = entropy(P[Y==0] / np.sum(P[Y==0]))
H_X_Y1 = entropy(P[Y==1] / np.sum(P[Y==1]))
H_X_given_Y = np.sum(P[Y==0])*H_X_Y0 + np.sum(P[Y==1])*H_X_Y1
MI = H_X - H_X_given_Y

print(f"\nH(X): {H_X:.4f} bits")
print(f"H(X|Y): {H_X_given_Y:.4f} bits")
print(f"Mutual Information I(X;Y): {MI:.4f} bits")
print(f"Percentage uncertainty removed: {(MI/H_X)*100:.2f}%")
```

# 2) Conditional density of the multivariate Gaussian

# Derivation of the Conditional Distribution of a Multivariate Normal Vector

We aim to derive the conditional distribution $f_{X|Y}(x \mid y)$ from the joint density $f_{X,Y}(z)$ of the concatenated random vector $Z = \begin{bmatrix} X \\ Y \end{bmatrix} \in \mathbb{R}^{n+m}$.


### 1. Definition and Block Matrix Structure

The joint density function is given by:
$$f_{X,Y}(z) = \frac{1}{(2\pi)^{\frac{n+m}{2}} |\Sigma|^{\frac{1}{2}}} \exp\left(-\frac{1}{2} (z-\mu)^T \Sigma^{-1} (z-\mu)\right)$$

where the partitioned mean and covariance matrix are:
$$\mu = \begin{bmatrix} \mu_X \\ \mu_Y \end{bmatrix}, \quad \Sigma = \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}$$

By definition, the conditional density is:
$$f_{X|Y}(x \mid y) = \frac{f_{X,Y}(x, y)}{f_Y(y)}$$

Since $Y$ is a marginal component of a multivariate normal vector, it is also normally distributed:
$$f_Y(y) = \frac{1}{(2\pi)^{\frac{m}{2}} |\Sigma_{YY}|^{\frac{1}{2}}} \exp\left(-\frac{1}{2} (y-\mu_Y)^T \Sigma_{YY}^{-1} (y-\mu_Y)\right)$$



### 2. Block Matrix Inversion & The Schur Complement

To evaluate the quadratic form in the joint density exponent, we invert the partitioned matrix $\Sigma$. Let $\Sigma_{X|Y}$ be the Schur complement of $\Sigma_{YY}$ in $\Sigma$:
$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$

Using the block matrix inversion theorem, the inverse covariance matrix is precisely partitioned as:
$$\Sigma^{-1} = \begin{bmatrix} \Sigma_{X|Y}^{-1} & -\Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \\ -\Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1} & \Sigma_{YY}^{-1} + \Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \end{bmatrix}$$

We can factorize the determinant of the joint matrix via the Schur complement property:
$$|\Sigma| = |\Sigma_{YY}| \cdot |\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}| = |\Sigma_{YY}| \cdot |\Sigma_{X|Y}|$$

Thus, the ratio of the normalization constants simplifies to:
$$\frac{\frac{1}{(2\pi)^{\frac{n+m}{2}} |\Sigma|^{\frac{1}{2}}}}{\frac{1}{(2\pi)^{\frac{m}{2}} |\Sigma_{YY}|^{\frac{1}{2}}}} = \frac{1}{(2\pi)^{\frac{n}{2}} |\Sigma_{X|Y}|^{\frac{1}{2}}}$$



### 3. Expansion of the Quadratic Form

Let $\tilde{x} = x - \mu_X$ and $\tilde{y} = y - \mu_Y$. The joint quadratic form inside the exponent can be expanded using the precise block components of $\Sigma^{-1}$[cite: 2]:

$$Q(x,y) = \begin{bmatrix} \tilde{x} \\ \tilde{y} \end{bmatrix}^T \Sigma^{-1} \begin{bmatrix} \tilde{x} \\ \tilde{y} \end{bmatrix}$$

$$Q(x,y) = \tilde{x}^T \Sigma_{X|Y}^{-1} \tilde{x} - \tilde{x}^T \Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} - \tilde{y}^T \Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}\tilde{x} + \tilde{y}^T \left( \Sigma_{YY}^{-1} + \Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \right) \tilde{y}$$

We can collect and factorize all terms containing the common matrix factor $\Sigma_{X|Y}^{-1}$:
$$\tilde{x}^T \Sigma_{X|Y}^{-1} \tilde{x} - \tilde{x}^T \Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} - \tilde{y}^T \Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}\tilde{x} + \tilde{y}^T \Sigma_{YY}^{-1}\Sigma_{YX}\Sigma_{X|Y}^{-1}\Sigma_{XY}\Sigma_{YY}^{-1} \tilde{y} = \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)^T \Sigma_{X|Y}^{-1} \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)$$

What remains outside of that factorization block is the lone $\tilde{y}$ expression:
$$\tilde{y}^T \Sigma_{YY}^{-1} \tilde{y}$$

Therefore, the total joint quadratic form beautifully simplifies to:
$$Q(x,y) = \tilde{y}^T \Sigma_{YY}^{-1} \tilde{y} + \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)^T \Sigma_{X|Y}^{-1} \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)$$



### 4. Constructing the Conditional Density

The conditional density calculation effectively subtracts the marginal quadratic form of $Y$ ($\tilde{y}^T \Sigma_{YY}^{-1} \tilde{y}$) from the total joint quadratic form $Q(x,y)$ within the exponential function:

$$-\frac{1}{2}Q(x,y) - \left( -\frac{1}{2}\tilde{y}^T \Sigma_{YY}^{-1} \tilde{y} \right) = -\frac{1}{2} \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)^T \Sigma_{X|Y}^{-1} \left( \tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} \right)$$

By substituting back our original vector translations ($\tilde{x} = x - \mu_X$ and $\tilde{y} = y - \mu_Y$), we can group the terms to define the conditional mean $\mu^*$:
$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$

This allows us to rewrite the inside of the quadratic vector structure as:
$$\tilde{x} - \Sigma_{XY}\Sigma_{YY}^{-1}\tilde{y} = (x - \mu_X) - \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y) = x - \mu^*$$

Bringing the simplified normalization constant block and the isolated exponent together reveals the exact probability density function of the conditional distribution:
$$f_{X|Y}(x \mid y) = \frac{1}{(2\pi)^{\frac{n}{2}}|\Sigma_{X|Y}|^{\frac{1}{2}}} \exp\left(-\frac{1}{2}(x - \mu^*)^T \Sigma_{X|Y}^{-1}(x - \mu^*)\right)$$

where:
$$\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$$
$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$

This matches the density expression for a multivariate normal distribution $\mathcal{N}(\mu^*, \Sigma_{X|Y})$.

```python
import numpy as np

def verify_conditional_gaussian():
    print("=== Verification of Joint Gaussian Partitioning ===")

    # Define arbitrary dimensions for X (n=3) and Y (m=2)
    n, m = 3, 2

    # 1. Define arbitrary mean vectors
    mu_X = np.array([1.0, 2.0, 3.0])
    mu_Y = np.array([-1.0, 0.5])

    # 2. Construct a valid symmetric, positive-definite covariance matrix
    np.random.seed(42)
    A = np.random.randn(n + m, n + m)
    Sigma = np.dot(A, A.T) # Assures positive definiteness

    # Slice the matrix into explicit block components
    Sigma_XX = Sigma[0:n, 0:n]
    Sigma_XY = Sigma[0:n, n:n+m]
    Sigma_YX = Sigma[n:n+m, 0:n]
    Sigma_YY = Sigma[n:n+m, n:n+m]

    # 3. Choose a specific observation value for y
    y_obs = np.array([0.0, 1.2])

    # 4. Compute the analytical conditional parameters
    Sigma_YY_inv = np.linalg.inv(Sigma_YY)

    # Conditional Mean: μ* = μ_X + Σ_XY * Σ_YY^-1 * (y - μ_Y)
    mu_star = mu_X + np.dot(Sigma_XY, np.dot(Sigma_YY_inv, (y_obs - mu_Y)))

    # Conditional Covariance (Schur Complement): Σ_X|Y = Σ_XX - Σ_XY * Σ_YY^-1 * Σ_YX
    Sigma_X_given_Y = Sigma_XX - np.dot(Sigma_XY, np.dot(Sigma_YY_inv, Sigma_YX))

    # 5. Verify the determinant identity property: |Σ| = |Σ_YY| * |Σ_X|Y|
    det_Sigma = np.linalg.det(Sigma)
    det_product = np.linalg.det(Sigma_YY) * np.linalg.det(Sigma_X_given_Y)

    # 6. Verify block inversion accuracy against full matrix inverse
    Sigma_inv_full = np.linalg.inv(Sigma)
    top_left_block_inv = np.linalg.inv(Sigma_X_given_Y)

    print("\n--- Structural Dimensions ---")
    print(f"X dimension (n): {n}, Y dimension (m): {m}")
    print(f"Full Joint Vector Matrix Shape: {Sigma.shape}")

    print("\n--- Determinant Identity Factorization Check ---")
    print(f"Determinant of full matrix |Σ| : {det_Sigma:.6f}")
    print(f"Factored product |Sigma_YY| * |Sigma_X|Y|: {det_product:.6f}")
    print(f"Identity Check Match: {np.isclose(det_Sigma, det_product)}")

    print("\n--- Matrix Inversion Component Identity Check ---")
    print("Comparing top-left block of full inverted matrix to inverse Schur complement:")
    print("Is matching component close?", np.allclose(Sigma_inv_full[0:n, 0:n], top_left_block_inv))

    print("\n--- Computed Conditional Parameters ---")
    print("Conditional Mean Vector (μ*):")
    print(mu_star)
    print("\nConditional Covariance Matrix (Σ_X|Y):")
    print(Sigma_X_given_Y)

if __name__ == '__main__':
    verify_conditional_gaussian()
```

# 3) continuous fault monitoring of a spindle motor


### 1. Joint Distribution

**Distribution of $(X, Y)$:**
Since $X$ and $Q$ are independent normal random variables, any linear combination of them is also normally distributed. Thus, $X$ and $Y = X + Q$ are jointly Gaussian. A multivariate normal distribution is completely specified by its mean vector and covariance matrix.
* Mean of $X$: $\mu_X = 0$
* Mean of $Y$: $\mu_Y = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$

Thus, the mean vector is:
$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left(\begin{bmatrix} 0 \\ 0 \end{bmatrix}, \Sigma\right)$$

**Showing $Y \sim \mathcal{N}(0, 5)$:**
Since $Y$ is a linear combination of independent normal variables, it is normal. Its mean is $\mu_Y = 0$. Its variance is:
$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) + 2\text{Cov}(X, Q)$$
Because $X$ and $Q$ are independent, $\text{Cov}(X, Q) = 0$.
$$\text{Var}(Y) = 1 + 4 = 5$$
Therefore, $Y \sim \mathcal{N}(0, 5)$.

**Covariance $\text{Cov}(X, Y)$:**
$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q) = \text{Var}(X) + 0 = 1$$

The full joint covariance matrix $\Sigma$ is:
$$\Sigma = \begin{bmatrix} \text{Var}(X) & \text{Cov}(X,Y) \\ \text{Cov}(X,Y) & \text{Var}(Y) \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix}$$



### 2. Information Geometry & Measurability

**Why $X$ is not $\sigma(Y)$-measurable:**
A random variable $X$ is measurable with respect to the $\sigma$-algebra generated by $Y$ (written as $X \in \sigma(Y)$) if and only if $X$ can be expressed as a deterministic function of $Y$ (i.e., $X = f(Y)$).
In this system, $Y = X + Q$. For any given value $Y = y$, there are infinitely many pairs $(x, q)$ such that $x + q = y$. Because the mapping from $Y$ to $X$ is one-to-many due to the stochastic noise $Q$, $X$ cannot be recovered uniquely or deterministically from $Y$. Hence, $X \notin \sigma(Y)$.

**Practical Interpretation ($Y = 5$):**
If the power meter reads $Y = 5$, this total power could be caused by:
1. High thermal load: e.g., $X = 4$ (severe overheating) and $Q = 1$ (normal cutting load).
2. High mechanical load: e.g., $X = 0$ (normal temperature) and $Q = 5$ (very hard material or heavy feed rate).

Because the sensor conflates both physical effects into a single scalar value, the engineer cannot isolate the source of the high reading from $Y$ alone.



### 3. Conditional Expectation (The Best Estimate)

Using the linear regression property for joint normal distributions:
$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$

Substituting our known parameters ($\mu_X = 0, \mu_Y = 0, \text{Cov}(X,Y) = 1, \text{Var}(Y) = 5$):
$$E[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$

**Signal-to-Total-Variance Ratio:**
We can write this general formula as:
$$E[X \mid Y] = \frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2} Y$$
Here, $\sigma_X^2 = 1$ is the "signal" variance (the feature we want to track) and $\sigma_X^2 + \sigma_Q^2 = 5$ is the total observed variance. The scaling factor $\frac{1}{5}$ directly reflects this ratio.


### 4. Verification of the Residual

Let $\hat{X} = E[X \mid Y] = \frac{1}{5}Y$. The residual is $Z = X - \hat{X} = X - \frac{1}{5}Y$.

**Proof that $E[Z] = 0$:**
$$E[Z] = E\left[X - \frac{1}{5}Y\right] = E[X] - \frac{1}{5}E[Y] = 0 - \frac{1}{5}(0) = 0$$
*(Alternatively, by the law of total expectation: $E[Z] = E[X - E[X|Y]] = E[X] - E[X] = 0$.)*

**Verification of Orthogonality ($\text{Cov}(Z, Y) = 0$):**
$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)$$
$$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \frac{1}{5}\text{Var}(Y)$$
Substituting $\text{Cov}(X, Y) = 1$ and $\text{Var}(Y) = 5$:
$$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$
Since the covariance is 0 and $E[Z]=0$, the error $Z$ is uncorrelated with the observation $Y$. This confirms the Orthogonality Principle.



### 5. Information Theory: Differential Entropy

The differential entropy of a 1D normal distribution $N(\mu, \sigma^2)$ is given by:
$$h = \frac{1}{2} \ln(2\pi e \sigma^2)$$

**Differential Entropy $h(X)$:**
Since $X \sim \mathcal{N}(0, 1)$, $\sigma_X^2 = 1$:
$$h(X) = \frac{1}{2} \ln(2\pi e \cdot 1) = \frac{1}{2} \ln(2\pi e) \approx 1.4189 \text{ nats}$$

**Conditional Differential Entropy $h(X \mid Y)$:**
First, find the conditional variance $\sigma^2_{X|Y}$:
$$\sigma^2_{X|Y} = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.2 = 0.8$$
Since $(X \mid Y=y)$ is normally distributed with a variance that does not depend on the specific value of $y$, the conditional entropy is:
$$h(X \mid Y) = \frac{1}{2} \ln(2\pi e \sigma^2_{X|Y}) = \frac{1}{2} \ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$


### 6. Mutual Information

**Calculate $I(X; Y)$:**
$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \ln(2\pi e) - \frac{1}{2} \ln(2\pi e \cdot 0.8)$$
$$I(X; Y) = \frac{1}{2} \ln\left(\frac{2\pi e}{2\pi e \cdot 0.8}\right) = \frac{1}{2} \ln\left(\frac{1}{0.8}\right) = \frac{1}{2} \ln(1.25) \approx 0.1116 \text{ nats}$$

**Percentage of Uncertainty Removed:**
The uncertainty of $X$ before observing $Y$ can be measured by its variance or via information metrics. In terms of variance reduction:
$$\text{Uncertainty Reduction} = \frac{\sigma_X^2 - \sigma^2_{X|Y}}{\sigma_X^2} = \frac{1 - 0.8}{1} = 20\%$$
Thus, observing the power meter removes **20%** of the variance uncertainty regarding the internal temperature.

**Effect of Noise Variance Increasing to 100:**
If $\text{Var}(Q) = 100$:
* $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 100 = 101$
* $\text{Cov}(X, Y) = 1$
* New conditional variance: $\sigma^2_{X|Y} = 1 - \frac{1^2}{101} = \frac{100}{101} \approx 0.9901$
* New Mutual Information: $I(X; Y) = \frac{1}{2} \ln\left(\frac{1}{0.9901}\right) = \frac{1}{2} \ln\left(\frac{101}{100}\right) \approx 0.00497 \text{ nats}$
* Percentage of uncertainty removed: $\frac{1 - 0.9901}{1} \approx 0.99\%$

*Interpretation:* As mechanical noise increases drastically, the power meter becomes dominated by material variation, making it almost completely uninformative about internal temperature.


### 7. Diagnostic Interpretation

**Why $Y$ is a Biased Estimator of $X$ given $Y=y$:**
The manager suggests using $\hat{X}_{manager} = Y$ as the estimate. Let's look at the expected value of this estimator *conditioned* on a specific observation $Y=y$:
$$E[\hat{X}_{manager} \mid Y=y] = E[Y \mid Y=y] = y$$
However, the true expected value of $X$ given $Y=y$ is $\frac{1}{5}y$. Therefore:
$$\text{Bias}(\hat{X}_{manager} \mid Y=y) = E[Y \mid Y=y] - E[X \mid Y=y] = y - \frac{1}{5}y = \frac{4}{5}y$$
Except when $y=0$, using $Y$ directly systematically overestimates $X$ when $y > 0$ and underestimates it when $y < 0$.

**Why $E[X \mid Y]$ "Shrinks" Toward the Mean:**
The optimal MMSE (Minimum Mean Squared Error) estimator is $E[X \mid Y] = \frac{1}{5}Y$. The factor $\frac{1}{5}$ acts as a **shrinkage factor**.
Because $Y$ contains a mix of both signal ($X$) and substantial noise ($Q$), a large value of $Y$ (e.g., $Y=5$) is statistically much more likely to be driven by a large noise event from $Q$ (which has a wider variance of 4) than an extreme temperature anomaly from $X$ (which has a narrow variance of 1).
The conditional expectation dynamically balances this by attenuating ("shrinking") the raw measurement toward the prior mean (0), protecting the system from generating false alarms due to random mechanical load spikes.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

def run_simulation(n_samples=100000):
    print("=== CNC Spindle Health Monitoring Simulation ===")

    # 1. Generate Data according to the setup
    # X ~ N(0, 1) - Internal Temperature
    X = np.random.normal(0, 1, n_samples)
    # Q ~ N(0, 4) - Mechanical Load Noise (Std dev = 2)
    Q = np.random.normal(0, 2, n_samples)
    # Y = X + Q - Observed Total Spindle Power
    Y = X + Q

    # 2. Empirical Verification of Part 1 & Part 3
    var_Y_empirical = np.var(Y)
    cov_XY_empirical = np.cov(X, Y)[0, 1]

    print(f"Empirical Var(Y): {var_Y_empirical:.4f} (Theoretical: 5.0)")
    print(f"Empirical Cov(X, Y): {cov_XY_empirical:.4f} (Theoretical: 1.0)")

    # 3. Conditional Expectation & Residuals (Part 4)
    # Best Estimate: E[X|Y] = 1/5 * Y
    X_hat = (1.0 / 5.0) * Y
    Z = X - X_hat

    mean_Z = np.mean(Z)
    cov_ZY = np.cov(Z, Y)[0, 1]

    print("\n--- Verification of the Residual ---")
    print(f"Mean of error Z: {mean_Z:.4f} (Theoretical: 0.0)")
    print(f"Cov(Z, Y) (Orthogonality): {cov_ZY:.4f} (Theoretical: 0.0)")

    # 5 & 6. Information Theory Metrics
    h_X = 0.5 * np.log(2 * np.pi * np.e * 1)
    sigma2_X_given_Y = 1.0 - (1.0**2 / 5.0) # 0.8
    h_X_given_Y = 0.5 * np.log(2 * np.pi * np.e * sigma2_X_given_Y)
    I_XY = h_X - h_X_given_Y
    pct_removed = (1.0 - sigma2_X_given_Y) * 100

    print("\n--- Information Theory Metrics ---")
    print(f"Differential Entropy h(X): {h_X:.4f} nats")
    print(f"Conditional Entropy h(X|Y): {h_X_given_Y:.4f} nats")
    print(f"Mutual Information I(X; Y): {I_XY:.4f} nats")
    print(f"Uncertainty (variance) removed: {pct_removed:.1f}%")

    # High noise scenario (Var(Q) = 100)
    Q_high = np.random.normal(0, 10, n_samples)
    Y_high = X + Q_high
    sigma2_X_given_Y_high = 1.0 - (1.0**2 / 101.0)
    pct_removed_high = (1.0 - sigma2_X_given_Y_high) * 100
    print(f"Uncertainty removed if Var(Q)=100: {pct_removed_high:.2f}%")

    # 7. Visualization of Shrinkage and Manager's Bias
    plt.figure(figsize=(10, 6))

    # Plot a subset of data points for clarity
    plt.scatter(Y[:1000], X[:1000], alpha=0.3, color='gray', label='Simulated Data (subset)')

    # Plot estimators
    y_range = np.linspace(-6, 6, 500)
    x_hat_opt = (1.0 / 5.0) * y_range
    x_hat_manager = y_range

    plt.plot(y_range, x_hat_opt, color='blue', linewidth=2.5, label='Optimal Estimate E[X|Y] = Y/5 (Shrinkage)')
    plt.plot(y_range, x_hat_manager, color='red', linestyle='--', linewidth=2, label="Manager's Estimate = Y (Biased)")

    # Highlight specific observation Y = 5
    plt.axvline(x=5, color='green', linestyle=':', label='Observation Y = 5')
    plt.scatter([5], [1], color='blue', s=100, zorder=5)
    plt.scatter([5], [5], color='red', s=100, zorder=5)

    plt.title('Optimal MMSE Estimation vs Manager Estimate (Shrinkage Effect)')
    plt.xlabel('Observed Total Spindle Power (Y)')
    plt.ylabel('Hidden Temperature Deviation (X)')
    plt.xlim(-6, 6)
    plt.ylim(-4, 6)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.savefig('spindle_estimation_analysis.png', dpi=300)
    print("\nPlot saved as 'spindle_estimation_analysis.png'.")

if __name__ == '__main__':
    run_simulation()
```
