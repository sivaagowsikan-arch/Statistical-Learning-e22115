# Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)

## 1. Formal Derivation of $\hat{\Sigma}_{MLE}$ Bias & Bessel's Correction

  The raw Maximum Likelihood Estimator (MLE) for the covariance matrix is defined as:
$$\hat{\Sigma}_{MLE} = \frac{1}{n}\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T}$$

We can expand the term inside the summation by adding and subtracting the true population mean $\mu$:
$$X_i - \hat{\mu}_n = (X_i - \mu) - (\hat{\mu}_n - \mu)$$

Substituting this back into the summation:
$$\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} = \sum_{i=1}^{n} \left[ (X_i - \mu) - (\hat{\mu}_n - \mu) \right] \left[ (X_i - \mu) - (\hat{\mu}_n - \mu) \right]^T$$

Expanding the terms gives:
$$\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} = \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - \sum_{i=1}^{n}(X_i - \mu)(\hat{\mu}_n - \mu)^T - \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(X_i - \mu)^T + \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$

Since $\sum_{i=1}^n (X_i - \mu) = n(\hat{\mu}_n - \mu)$[cite: 11], the middle cross-product terms simplify:
$$\sum_{i=1}^{n}(X_i - \mu)(\hat{\mu}_n - \mu)^T = n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$

This allows the full summation identity to reduce cleanly to:
$$\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} = \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$

Taking the mathematical expectation on both sides:
$$\mathbb{E}\left[ \sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} \right] = \sum_{i=1}^{n}\mathbb{E}[(X_i - \mu)(X_i - \mu)^T] - n\mathbb{E}[(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T]$$

  By definition, $\mathbb{E}[(X_i - \mu)(X_i - \mu)^T] = \Sigma$. Since the sample mean has variance $\text{Var}(\hat{\mu}_n) = \frac{1}{n}\Sigma$, we get:
$$\mathbb{E}\left[ \sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} \right] = n\Sigma - n\left(\frac{1}{n}\Sigma\right) = (n-1)\Sigma$$

Now, substitute this back into the expectation of the raw MLE estimator:
$$\mathbb{E}[\hat{\Sigma}_{MLE}] = \mathbb{E}\left[ \frac{1}{n}\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T} \right] = \frac{n-1}{n}\Sigma$$
  This formally proves that $\hat{\Sigma}_{MLE}$ is inherently biased in finite samples.

### Mathematically Applying Bessel's Correction
  To eliminate this finite-sample bias, we multiply $\hat{\Sigma}_{MLE}$ by the correction factor $\frac{n}{n-1}$:
$$S = \frac{n}{n-1}\hat{\Sigma}_{MLE} = \frac{1}{n-1}\sum_{i=1}^{n}(X_{i}-\hat{\mu}_{n})(X_{i}-\hat{\mu}_{n})^{T}$$
  Taking its expectation yields the unique, unbiased estimator $S$:
$$\mathbb{E}[S] = \frac{n}{n-1}\mathbb{E}[\hat{\Sigma}_{MLE}] = \frac{n}{n-1} \left( \frac{n-1}{n}\Sigma \right) = \Sigma$$

## 2. Errors and Thresholds in Continuous Structural Health Monitoring (SHM)

*   **Type I Error ($\alpha$):** A false alarm.   This occurs when the diagnostic layer falsely concludes that a healthy structure is damaged, triggering unnecessary operational shutdowns.
*   **Type II Error ($\beta$):** A missed detection.   This occurs when a true internal structural defect or drift is present, but the diagnostic system fails to detect it, 17].

### Mathematical Consequence of $\alpha = 0.0001$
  Setting an ultra-conservative significance threshold ($\alpha = 0.0001$) minimizes false alarms.   However, because $\alpha$ and $\beta$ are inversely related, this choice severely reduces the statistical power ($1 - \beta$) of your diagnostic layer.   The framework becomes heavily desensitized and may completely miss early-stage asset degradation.

### Geometric Volume of the Confidence Ellipsoid
  This decision drastically **inflates the geometric volume** of your "healthy operation" confidence ellipsoid.   The acceptance boundary expands outward, meaning a data point must represent an extreme operational anomaly before it trips the alarm system.

# Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)

## 1. Rigorous Statement of Slutsky's Theorem
Let $X_n$ and $Y_n$ be sequences of random vectors or matrices.   If $X_n$ converges in distribution to a random element $X$ ($X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a constant matrix $c$ ($Y_n \xrightarrow{p} c$), then:
1. $X_n + Y_n \xrightarrow{d} X + c$
2. $Y_n X_n \xrightarrow{d} cX$
3. $Y_n^{-1} X_n \xrightarrow{d} c^{-1}X$ (given $c$ is invertible)

## 2. Justification for Substituting $S$ for $\Sigma$
  The Multivariate Central Limit Theorem establishes that as $n\rightarrow\infty$:
$$\sqrt{n}(\hat{\mu}_{n}-\mu)\xrightarrow{d}\mathcal{N}(0,\Sigma)$$

  The true population covariance matrix $\Sigma$ is latent and unobservable.   However, by the Weak Law of Large Numbers, the empirical sample covariance matrix $S$ is a consistent estimator of $\Sigma$, meaning:
$$S \xrightarrow{p} \Sigma \implies S^{-1/2} \xrightarrow{p} \Sigma^{-1/2} \quad \text{as } n \to \infty$$

  By assigning $X_n = \sqrt{n}(\hat{\mu}_{n}-\mu)$ and $Y_n = S^{-1/2}$, we invoke Slutsky's multiplicative property
$$S^{-1/2} \cdot \sqrt{n}(\hat{\mu}_{n}-\mu) \xrightarrow{d} \Sigma^{-1/2} \cdot \mathcal{N}(0, \Sigma) = \mathcal{N}(0, I)$$

  Since the asymptotic distribution is now free of the latent parameter $\Sigma$, we can reorganize the expression for large finite samples:
$$\sqrt{n}(\hat{\mu}_{n}-\mu) \approx \mathcal{N}(0, S) \implies \hat{\mu}_{n}\sim\mathcal{N}\left(\mu,\frac{1}{n}S\right)$$
  This formally operationalizes the practical parametric distribution used in engineering monitoring pipelines.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# DO NOT MODIFY: Synthetic Strain Sensor Data Generator
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)
# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    X = df.to_numpy()
    n, m = X.shape
    chunk_size = n // g_chunks

    # 1. Compute Global Mean
    global_mean = np.mean(X, axis=0)

    # Initialize Matrices
    W = np.zeros((m, m))
    B = np.zeros((m, m))

    # 2. Extract Chunks and Calculate Within (W) & Between (B) Matrices
    for g in range(g_chunks):
        chunk = X[g * chunk_size : (g + 1) * chunk_size, :]
        chunk_mean = np.mean(chunk, axis=0)
        n_j = len(chunk)

        # Within scatter matrix contribution
        chunk_centered = chunk - chunk_mean
        W += chunk_centered.T @ chunk_centered

        # Between scatter matrix contribution
        mean_diff = (chunk_mean - global_mean).reshape(-1, 1)
        B += n_j * (mean_diff @ mean_diff.T)

    # 3. Compute Wilks' Lambda
    wilks_lambda = np.linalg.det(W) / np.linalg.det(B + W)

    # 4. Transform using Bartlett's Chi-Square approximation
    df_chi = m * (g_chunks - 1)
    bartlett_chi2 = - (n - 1 - (m + g_chunks) / 2.0) * np.log(wilks_lambda)
    p_value = stats.chi2.sf(bartlett_chi2, df_chi)

    return {
        "Wilks' Lambda": float(wilks_lambda),
        "Bartlett Chi2": float(bartlett_chi2),
        "p-value": float(p_value),
        "Degrees of Freedom": int(df_chi)
    }

# Execute completed framework
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
print(results)
```

# PCA & FA Theoretical Fundamentals

## 1. PCA Coordinate Transformations & Orthogonality
  Given the orthogonal projection mapping $Z_{i}=P^{T}\tilde{X}_{i}$ and $\Sigma = \mathbb{E}[\tilde{X}_{i}\tilde{X}_{i}^{T}] = P\Lambda P^{T}$:
$$\mathbb{E}[Z_{i}Z_{i}^{T}] = \mathbb{E}[(P^{T}\tilde{X}_{i})(P^{T}\tilde{X}_{i})^{T}] = P^{T}\mathbb{E}[\tilde{X}_{i}\tilde{X}_{i}^{T}]P = P^{T}\Sigma P$$
Substituting the spectral decomposition:
$$\mathbb{E}[Z_{i}Z_{i}^{T}] = P^{T}(P\Lambda P^{T})P$$
  Since $P$ is an orthogonal matrix ($P^{T}P = I$):
$$\mathbb{E}[Z_{i}Z_{i}^{T}] = I \Lambda I = \Lambda$$

### Geometric & Statistical Meaning
  Statistically, because $\Lambda$ is strictly diagonal  , all cross-covariances between distinct coordinates are zero ($\text{Cov}(Z_{i,j}, Z_{i,k}) = 0$ for $j \neq k$).   This means PCA completely de-correlates the incoming sensor streams.   Geometrically, it reflects a rigid rotation of the feature space coordinate axes to align perfectly with the orthogonal directions of maximum data variance.

## 2. Total Variance Conservation Law
  Utilizing the cyclic permutation property of the trace operator ($\text{tr}(ABC) = \text{tr}(CAB)$):
$$\text{tr}(\Lambda) = \text{tr}(P^{T}\Sigma P) = \text{tr}(P P^{T}\Sigma)$$
  Because $P$ is orthogonal ($P P^{T} = I$), this simplifies directly to:
$$\text{tr}(\Lambda) = \text{tr}(\Sigma) = \sum_{j=1}^{m}\lambda_{j}$$
  This proves that the total system variance is preserved during the mapping from raw space to principal space.

*   **Cumulative Explained Variance Ratio:** $\Phi(k) = \frac{\sum_{j=1}^{k}\lambda_{j}}{\sum_{j=1}^{m}\lambda_{j}}$
*   **Residual Unexplained Variance Ratio:** $\Psi(k) = \frac{\sum_{j=k+1}^{m}\lambda_{j}}{\sum_{j=1}^{m}\lambda_{j}}$

## 3. Fundamental Equation of Factor Analysis
  Given $Z_{i} = \Lambda F_{i} + \epsilon_{i}$ where $\mathbb{E}[F_{i}F_{i}^{T}] = I$, $\mathbb{E}[\epsilon_{i}\epsilon_{i}^{T}] = \Psi$, and $\mathbb{E}[\epsilon_{i}F_{i}^{T}] = 0$
$$R = \mathbb{E}[Z_{i}Z_{i}^{T}] = \mathbb{E}[(\Lambda F_{i} + \epsilon_{i})(\Lambda F_{i} + \epsilon_{i})^{T}]$$
$$R = \Lambda\mathbb{E}[F_{i}F_{i}^{T}]\Lambda^{T} + \Lambda\mathbb{E}[F_{i}\epsilon_{i}^{T}] + \mathbb{E}[\epsilon_{i}F_{i}^{T}]\Lambda^{T} + \mathbb{E}[\epsilon_{i}\epsilon_{i}^{T}]$$
  Substituting the known model identities yields:
$$R = \Lambda(I)\Lambda^{T} + 0 + 0 + \Psi = \Lambda\Lambda^{T} + \Psi$$

### Physical Interpretations for Diagnostics
*   **Communality ($h_{j}^{2}$):** The portion of a sensor's variance driven by the core underlying physical forces (e.g., global structural loads or thermal stress) affecting the asset.
* **Uniqueness ($\varphi_{j}^{2}$):** The isolated variance specific to that single hardware channel, representing local electrical noise or point-level sensor degradation.

$$\Lambda^* (\Lambda^*)^T = (\Lambda T)(T^T \Lambda^T) = \Lambda (T T^T) \Lambda^T = \Lambda \Lambda^T$$
Because the matrix product remains identical, the diagonal elements ($h_j^2$) and the global covariance approximation remain completely unchanged under rotation.
